```python
%load_ext autoreload
%autoreload 2

import matplotlib.pyplot as plt
from PIL import Image
import numpy as np
import torch
import torchvision
```

```python
import warnings
warnings.filterwarnings('ignore')
import sys
sys.path.append("..")
from dar_package.datasets.vaihingen import VaihingenDataset
from dar_package.datasets.bing import BingDataset
```

```python
dataset = VaihingenDataset('val_1', normalize=True)
sample = dataset[0]
im, mask, interp_coords, dist_tr_inside, origin = (
    sample['image'], sample['semantic_mask'], sample['interp_poly'],
    sample['distance_transform'], sample['init_contour_origin'])
im = dataset.unnormalize(im)
im = torchvision.transforms.functional.to_pil_image(im)
mask = mask.numpy().squeeze()

plt.figure(figsize=(10,10))
f, ax = plt.subplots(1, 3, figsize=(10, 5))
ax[0].imshow(im)
ax[0].scatter(origin[0], origin[1], color='red')
ax[0].scatter(interp_coords[:, 0], interp_coords[:, 1], color='cyan', s=3)
im = ax[1].imshow(mask)
f.colorbar(im, ax=ax[1])
im = ax[2].imshow(dist_tr_inside)
f.colorbar(im, ax=ax[2])
```
